### Part 1: Python List Operations

**Objective**: Practice writing a Python function to process a list of numbers.

**Scenario**: You have a list of daily temperature readings (in Celsius) for a week. You need to identify days where the temperature was above a certain threshold and calculate their average.

**Instructions**:
1.  **Run the pre-code block below** to define the `daily_temperatures` list and `temperature_threshold`.
2.  **Create a Python function** named `analyze_temperatures` that takes two arguments: a list of temperatures and a threshold temperature.
3.  **Inside the function**:
    *   Identify all temperatures that are *strictly greater* than the given threshold.
    *   If there are any temperatures above the threshold, calculate their average.
    *   Return the list of above-threshold temperatures and their average. If no temperatures are above the threshold, return an empty list and `None` for the average.
4.  **Call your function** with the provided `daily_temperatures` and `temperature_threshold`.
5.  **Print** the results in a user-friendly format.

```python
# Pre-code for Part 1
daily_temperatures = [22, 25, 19, 28, 23, 30, 26] # Temperatures in Celsius
temperature_threshold = 24
```

```python
# Your code for Part 1 goes here

def analyze_temperatures(temperatures, threshold):
    # TODO: Implement the function as per instructions

    above_threshold = []

    for temp in temperatures:
        if temp > threshold:
            above_threshold.append(temp)

    if len(above_threshold) > 0:
        average = sum(above_threshold) / len(above_threshold)
    else:
        average = None

    return above_threshold, average
    pass

# TODO: Call the function and print results
above_temps, avg_temp = analyze_temperatures(
    daily_temperatures,
    temperature_threshold
)

print("Temperatures above threshold:", above_temps)
print("Average temperature:", avg_temp)
```

### Part 3: Bonus Problem - Object-Oriented Book Management

**Objective**: Practice object-oriented programming concepts (`class`, `__init__`) by creating a `Book` class and performing operations on a collection of books.

**Scenario**: You are managing a small library and want to organize book information using Python classes.

**Instructions**:
1.  **Define a `Book` class** with an `__init__` method that takes `title` and `author` as arguments and stores them as instance attributes.
2.  **Create a list of `Book` objects** using the provided `book_data`.
3.  **Implement a method** within the `Book` class (or a separate function) to:
    *   Find the book with the **longest title**.
    *   List all books written by a specific author (e.g., 'George Orwell').
4.  **Print** the results in a clear and readable format.

```python
# Pre-code for Part 3
book_data = [
    {'title': '1984', 'author': 'George Orwell'},
    {'title': 'Brave New World', 'author': 'Aldous Huxley'},
    {'title': 'Animal Farm', 'author': 'George Orwell'},
    {'title': 'The Great Gatsby', 'author': 'F. Scott Fitzgerald'},
    {'title': 'To Kill a Mockingbird', 'author': 'Harper Lee'}
]


# Your code for Part 3 goes here

# TODO: Define the Book class

class Book:

    def __init__(self, title, author):
        self.title = title
        self.author = author


# TODO: Create Book objects and perform operations

books = []

for item in book_data:
    books.append(Book(item["title"], item["author"]))

# Find the book with the longest title
longest_book = max(books, key=lambda book: len(book.title))

# Find books written by George Orwell
orwell_books = []

for book in books:
    if book.author == "George Orwell":
        orwell_books.append(book)


# TODO: Print results

print("Book with the longest title:")
print(longest_book.title, "-", longest_book.author)

print("\nBooks by George Orwell:")

for book in orwell_books:
    print(book.title)
```
